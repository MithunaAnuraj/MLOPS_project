```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

```python
ratings = pd.read_csv('ml-latest-small/ml-latest-small/ratings.csv')
ratings.head()
```

```python
movie_meta = pd.read_csv('ml-latest-small/ml-latest-small/movies.csv')
movie_meta.head()
```

```python
ratings_movie_title = pd.merge(left=ratings,right = movie_meta,how='left',left_on = 'movieId',right_on='movieId')
ratings_movie_title.drop(['timestamp','genres'],inplace=True,axis=1)
ratings_movie_title.head()
```

```python

movie_ratings_pivot = ratings_movie_title.pivot(index=['title','movieId'],columns='userId',values='rating').fillna(0)
movie_ratings_pivot.head()
```

```python
movie_ratings_pivot = movie_ratings_pivot.droplevel(level='movieId')
movie_ratings_pivot.head()
```
